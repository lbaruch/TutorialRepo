```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Load datasets
user_usage = pd.read_csv('user_usage.csv')
user_device = pd.read_csv('user_device.csv')
devices = pd.read_csv('android_devices.csv')

# Print the DataFrames one after another
print("User Usage DataFrame:")
display(user_usage)

print("\nUser Device DataFrame:")
display(user_device)

print("\nDevices DataFrame:")
display(devices)
```

```python
#Display the DataFrames with good looking table

# Display the DataFrames
print("User Usage DataFrame:")
display(user_usage)  # Display user_usage DataFrame
print("\nUser Device DataFrame:")
display(user_device)  # Display user_device DataFrame
print("\nDevices DataFrame:")
display(devices)  # Display devices DataFrame

```

```python
# Display the Info of the DataFrames
print("\nUser Usage DataFrame:")
user_usage.info()  # Display user_usage DataFrame
print("\nUser Device DataFrame:")
user_device.info()  # Display user_device DataFrame
print("\nDevices DataFrame:")
devices.info()  # Display devices DataFrame
```

```python
# Handle missing data (fill Marketing Name with Device if missing)
devices['Model'].fillna(devices['Device'], inplace=True) #Model - Marketing Name
```

```python
print("\nDevices DataFrame:")
devices.info()  # Display devices DataFrame
```

```python
#Check the number of unique ID's in each of the charts
# Number of unique "use_id" in user_usage DataFrame
unique_user_usage_ids = user_usage['use_id'].nunique()

# Number of unique "use_id" in user_device DataFrame
unique_user_device_ids = user_device['use_id'].nunique()

# Number of unique "use_id" in devices DataFrame (if it exists in this dataset)
# If there is no "use_id" in this dataset, skip this line
unique_devices_ids = devices['Device'].nunique() if 'Device' in devices.columns else None

# Print the results
print(f"Unique 'use_id' in user_usage: {unique_user_usage_ids}")
print(f"Unique 'use_id' in user_device: {unique_user_device_ids}")
if unique_devices_ids is not None:
    print(f"Unique 'use_id' in android devices: {unique_devices_ids}")
else:
    print("'use_id' column not present in devices dataset.")
```

```python
# Merge user_usage with user_device on 'use_id'
usage_device_data = pd.merge(user_usage, user_device, on='use_id')
usage_device_data = pd.merge(user_usage,
                 user_device[['use_id', 'platform', 'device']],
                 on='use_id')
usage_device_data.head()
print("\nUsage_Device_Data Merged DataFrame:")
display(usage_device_data)  # Display devices DataFrame
display(usage_device_data.info())
```

```python
# Use Left Join Merge
usage_device_data = pd.merge(user_usage,
                 user_device[['use_id', 'platform', 'device']],
                 on='use_id',
                 how='left')



# We take the "device" column from one DF and "model" from other DF and merge them
devices.rename(columns={"Retail Branding": "manufacturer"}, inplace=True)
usage_device_data = pd.merge(usage_device_data, 
                  devices[['manufacturer', 'Model']],
                  left_on='device',
                  right_on='Model',
                  how='left')
display(usage_device_data.head())
```

```python
#First Insight that shows the diffrent Manufactures
#Now we do agregation using GroupBy
usage_device_data .groupby("manufacturer").agg({
        "outgoing_mins_per_month": "mean",
        "outgoing_sms_per_month": "mean",
        "monthly_mb": "mean",
        "use_id": "count"
    })
```

```python
# Write the final merged DataFrame to a new CSV file
usage_device_data.to_csv('final_merged_data.csv', index=False)

# Check if there are rows with platform 'IOS'
ios_rows = usage_device_data[usage_device_data['platform'] == 'ios']

# Display the rows with platform 'IOS'
display(ios_rows)
display(usage_device_data.info())
```

```python


# Insight 1: Total data usage by platform (iOS vs Android)
data_usage_platform = usage_device_data.groupby('platform')['monthly_mb'].sum().reset_index()
# Insight 2: Average outgoing minutes per device manufacturer
outgoing_mins_manufacturer = usage_device_data.groupby('device')['outgoing_mins_per_month'].mean().reset_index() #manufacturer - device

# Insight 3: SMS usage comparison between work and personal devices
sms_usage_type = usage_device_data.groupby('use_id')['outgoing_sms_per_month'].mean().reset_index()

# Plotting the insights
plt.figure(figsize=(14, 8))

# Insight 1 Visualization: Data Usage by Platform
plt.subplot(1, 3, 1)
sns.barplot(x='platform', y='monthly_mb', data=data_usage_platform)
plt.title('Total Data Usage by Platform')
plt.xlabel('Platform')
plt.ylabel('Total Monthly MB')

# Insight 2 Visualization: Outgoing Minutes by Manufacturer
plt.subplot(1, 3, 2)
sns.barplot(x='device', y='outgoing_mins_per_month', data=outgoing_mins_manufacturer)
plt.title('Average Outgoing Minutes by Manufacturer')
plt.xticks(rotation=45)
plt.xlabel('Manufacturer')
plt.ylabel('Avg Outgoing Minutes')

# Insight 3 Visualization: SMS Usage by Use Type (Work/Personal)
plt.subplot(1, 3, 3)
sns.barplot(x='use_id', y='outgoing_sms_per_month', data=sms_usage_type)
plt.title('Average SMS Usage by Device Type')
plt.xlabel('Use Type')
plt.ylabel('Avg Outgoing SMS')

plt.tight_layout()
plt.show()
```
